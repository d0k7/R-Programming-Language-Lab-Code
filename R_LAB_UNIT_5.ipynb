```r
#Q1) Create a database with the fields of weight,height and sex then create a plot of weight on the x-axis and height on the y-axis. Use different point characters or colors
#     to distinguish between males and females and provide a matching legend. Label the axes and give the plot a title.




# Create the database
my_data <- data.frame(
  weight = c(60, 70, 55, 65, 75, 68),
  height = c(160, 170, 155, 165, 175, 168),
  sex = c("Male", "Male", "Female", "Female", "Male", "Female")
)

# Plot using base R graphics
plot(my_data$weight, my_data$height, col = ifelse(my_data$sex == "Male", "blue", "red"),
     pch = ifelse(my_data$sex == "Male", 16, 17), xlab = "Weight", ylab = "Height",
     main = "Weight vs. Height", xlim = c(50, 80), ylim = c(150, 180))
legend("topleft", legend = c("Male", "Female"), col = c("blue", "red"), pch = c(16, 17))
```

```r
#Q2) create a plot using ggplot2 for the same database consists of weight on the x-axis and height on the y- axis. Use different point characters or colors to
#    distinguish between males and females and provide a matching legend. Label the axes and give the plot a title.




# Load ggplot2 library
library(ggplot2)

# Plot using ggplot2
ggplot(data = my_data, aes(x = weight, y = height, color = sex, shape = sex)) +
  geom_point(size = 4) +
  labs(x = "Weight", y = "Height", title = "Weight vs. Height") +
  scale_color_manual(values = c("blue", "red")) +
  scale_shape_manual(values = c(16, 17)) +
  theme_minimal()

```

```r
#Q3) Write R code that will plot education on the x-axis and income on the y-axis, with both x- and y-axis limits fixed to be [0;100]. Provide appropriate axis labels.
#    For jobs with a prestige value of less than or equal to 80, use a black * as the point character. For jobs with prestige greater than 80, use a blue @.




# Create the database
my_data <- data.frame(
  education = c(80, 90, 60, 70, 75, 85),
  income = c(50, 70, 40, 60, 55, 80),
  prestige = c(90, 100, 70, 80, 75, 95)
)

# Plot with different point characters based on prestige values
plot(my_data$education, my_data$income, col = ifelse(my_data$prestige <= 80, "black", "blue"),
     pch = ifelse(my_data$prestige <= 80, "*", "@"), xlab = "Education", ylab = "Income",
     main = "Education vs. Income", xlim = c(0, 100), ylim = c(0, 100))
```
